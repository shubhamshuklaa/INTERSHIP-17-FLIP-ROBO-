### 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
#### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
#### You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(5)

Rank = []
Name = []
Artist = []
Upload = []
Views = []
time.sleep(1)

sno = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
for i in sno:
    Rank.append(i.text) 
    
video = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
for i in video:
    Name.append((i.text)[:-4].replace('"',''))  
uploader = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
for i in uploader:
    Artist.append(i.text)  
    
date = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
for i in date:
    Upload.append(i.text)  
    
view = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
for i in view:
    Views.append(i.text)  
time.sleep(3)

#Creating a dataframe.
Youtube = pd.DataFrame({})

Youtube['Rank'] = Rank
Youtube['Name of the video'] = Name
Youtube['Artist'] = Artist
Youtube['Upload date'] = Upload
Youtube['Views (In billions)'] = Views
```

### 2. Scrape the details team India’s international fixtures from bcci.tv.
#### Url = https://www.bcci.tv/.
#### You need to find following details:
A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Time
#### Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
url = "https://www.bcci.tv/"
driver.get(url)
time.sleep(5)

link_to_page = driver.find_element_by_xpath('//li[@class="navigation__item  "][1]/div[2]/div/ul/li/a[1]').get_attribute('href')

driver.get(link_to_page) 
time.sleep(5)

title = []
series = []
place = []
date = []
time1 = []

days=[]
months=[]

title_tags = driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
for i in title_tags:
    title.append(i.text)  

place_tag = driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
for i in place_tag:
    place.append(i.text) 

series_tags = driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__format']")
for i in series_tags:
    series.append(i.text)  

time_tags = driver.find_elements_by_xpath("//span[@class='fixture__time']")
for i in time_tags:
    time1.append(i.text) 

day_tag = driver.find_elements_by_xpath("//span[@class='fixture__date']")
month_tag = driver.find_elements_by_xpath("//span[@class='fixture__month']")

for i in day_tag:
    days.append(i.text)
    
for i in month_tag:
    months.append(i.text)
    
for i in range(0, len(days)):
    date.append(days[i]+" "+months[i])  

#Creating a dataframe.
Matches = pd.DataFrame({})

Matches['Match Title'] = title
Matches['Series'] = series
Matches['Place'] = place
Matches['Date'] = date
Matches['Time'] = time1
```

#### Url = https://github.com/
#### You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
url = 'https://github.com/'
driver.get(url)
time.sleep(3)

trend = driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[4]/details/div/ul[2]/li[3]/a").get_attribute('href')

driver.get(trend)  
time.sleep(4)

title = []
desc=[]
count=[]
lang=[]

name = driver.find_elements_by_xpath("//div[@class='Box']/div[2]/article/h1/a")
for i in name:
    title.append((i.text).split('/')[1]) 
description = driver.find_elements_by_xpath("//article[@class='Box-row']")
for i in description:
    try:
        desc1 = i.find_element_by_xpath("./p")
        desc.append(desc1.text)
    except NoSuchElementException:
        desc.append('--')

contribute = driver.find_elements_by_xpath("//div[@class='Box']/div[2]/article/div[2]/a[2]")
for i in contribute:
    count.append(i.text)  

language= driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']")
for i in language:
    try :
        lang1 = i.find_element_by_xpath("./span/span[2]")
        lang.append(lang1.text)
    except NoSuchElementException:
        lang.append('--')  
#Creating a dataframe.
Github = pd.DataFrame({})

Github['Repository title'] = title
Github['Repository description'] = desc
Github['Contributors count'] = count
Github['Language used'] = lang
```

### 10. Details of Datasets from UCI machine learning repositories.
#### Url = https://archive.ics.uci.edu/
#### You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year

```python
url = 'https://archive.ics.uci.edu/'
driver.get(url)
time.sleep(5)

all_data = driver.find_element_by_xpath("//body/table[2]//a").get_attribute('href') 
driver.get(all_data)                                                                
time.sleep(10)

name=[]
type_=[]
task=[]
attribute=[]
noofinstance=[]
noofattribute=[]
year=[]

source = driver.find_elements_by_xpath("//table[@border='1']/tbody/tr")

## Extracted Names
for i in source:
    try:
        name1 = i.find_element_by_xpath("./td/table/tbody/tr/td[2]/p/b/a")
        name.append(name1.text)
    except:
        name.append("WRONG")
        

##Extracted Type
for i in source:
    try:
        type1 = i.find_element_by_xpath("./td[2]/p")
        type_.append(type1.text)
    except:
        type_.append("WRONG")

##Extracted Task
for i in source:
    try:
        task1 = i.find_element_by_xpath("./td[3]/p")
        task.append(task1.text)
    except:
        task.append("WRONG")

##Extracted Attribute
for i in source:
    try:
        attri = i.find_element_by_xpath("./td[4]/p")
        attribute.append(attri.text)
    except:
        attribute.append("WRONG")

##Extracted No of instances
for i in source:
    try:
        instance = i.find_element_by_xpath("./td[5]/p")
        noofinstance.append(instance.text)
    except:
        noofinstance.append("WRONG")

##Extracted No of attributes
for i in source:
    try:
        noattri = i.find_element_by_xpath("./td[6]/p")
        noofattribute.append(noattri.text)
    except:
        noofattribute.append("WRONG")

##Extracted year
for i in source:
    try:
        year1 = i.find_element_by_xpath("./td[7]/p")
        year.append(year1.text)
    except:
        year.append("WRONG")

#Creating a dataframe.
Datasets = pd.DataFrame({})

Datasets['Dataset name'] = name[1:]
Datasets['Data type'] = type_[1:]
Datasets['Task'] = task[1:]
Datasets['Attribute type'] = attribute[1:]
Datasets['No of instances'] = noofinstance[1:]
Datasets['No of attribute'] = noofattribute[1:]
Datasets['Year'] = year[1:]
```
